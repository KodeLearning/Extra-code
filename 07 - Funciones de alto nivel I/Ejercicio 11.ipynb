# Ejercicio 1

## Enunciado

Imagina la siguientes lista con 5 elementos (los que tú quieras). Crea una función que permita intercambiar dos elementos cualesquiera. Tomando como ejemplo la siguiente lista

```
lista = ['a', 'b', 'c', 'd', 'e']
```

La llamada a nuestra función `intercambia(lista, posicion_origen, posicion_destino)` con `posicion_origen = 0` y `posicion_destino = 3 `debería devolver la siguiente lista
```
['d', 'b', 'c', 'a', 'e]
```

Como puede verse se han intercambiado las posiciones 0 y 3 sin tocar las demás.

Debes construir la función `intercambia(lista, posicion_origen, posicion_destino)`


## Ayuda a la implementación

Al ser opcional, si no se necesita no se pone este apartado
