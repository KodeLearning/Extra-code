# Ejercicio 2. Filtrado de listas

## Enunciado

Crea una función que reciba dos parámetros de entrada, una lista y una función de filtrado.

La función debe devolver una nueva lista en la que solo figuren los elementos que cumplan con la función de filtrado.

## Ayuda a la implementación

La función de filtrado deberá equivaler a un predicado que devuelva True o False al procesar cada item de la lista.

Dos buenas formas de probar serían extraer todos los pares de una lista de números, o extraer todas las vocales de una palabra. En este último caso devolverá una lista, no una cadena.
