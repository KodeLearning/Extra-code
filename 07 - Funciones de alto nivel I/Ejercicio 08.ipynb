# Ejercicio 8. Composición de funciones (notas nombradas)

## Enunciado

Partimos ahora de la siguiente estructura de datos
```
alumnos = [
    ('María García', (('Matemáticas', 8), ('Lengua', 9), ('Inglés', 7), ('Conocimiento del medio', 8), ('Música', 6))),
    ('Juan Pérez', (('Matemáticas', 7), ('Lengua', 8), ('Inglés', 9), ('Conocimiento del medio', 6), ('Música', 7))),
    ('Lucía Martínez', (('Matemáticas', 9), ('Lengua', 7), ('Inglés', 8), ('Conocimiento del medio', 9), ('Música', 8))),
    ('Pedro Ruiz', (('Matemáticas', 6), ('Lengua', 7), ('Inglés', 8), ('Conocimiento del medio', 7), ('Música', 6))),
    ('Ana Sánchez', (('Matemáticas', 8), ('Lengua', 8), ('Inglés', 7), ('Conocimiento del medio', 7), ('Música', 9))),
    ('Carlos Gómez', (('Matemáticas', 9), ('Lengua', 7), ('Inglés', 6), ('Conocimiento del medio', 8), ('Música', 7)))
]
```

Procesa esta lista utilizando composición de funciones para obtener una lista de tuplas nueva que tenga la siguiente estructura.
```
[('María García', 7.6, 'Notable'), ...]
```

Como ves se ha añadido el nombre de la nota numérica. Corresponden a los siguientes criterios

- por debajo de 3: Muy Deficiente
- por debajo de 5: Insuficiente
- por debajo de 6: Suficiente
- por debajo de 7: Bien
- por debajo de 9: Notable
- por debajo de 10: Sobresaliente
- igual a 10: Matrícula de honor
