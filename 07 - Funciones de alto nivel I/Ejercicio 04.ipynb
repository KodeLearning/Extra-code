# Ejercicio 4. Transformando cadenas con lambdas

## Enunciado

Partiendo de la función de transformación de cadenas del ejercicio 1 de este bloque. Utiliza funciones anónimas para 
1. duplicar cada caracter
2. hacer mayúsculas las vocales y minúsculas el resto de caracteres
3. eliminar las letras h (son mudas)
4. cambiar las oes por caritas sonrientes (elige si emoticono o la secuencia `:)`
