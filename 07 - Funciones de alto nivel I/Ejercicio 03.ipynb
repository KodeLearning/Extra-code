# Ejercicio 3. Acumulado de listas

## Enunciado

Crea una función que procese todos los elementos de una lista y los reduzca a un valor. 

Imagina una lista de números, por ejemplo (2, 4, 6, 8, 10). Debes crear una función que tome todos esos valores y los multiplique. Para ello debes utilizar una función de alto nivel con la siguiente firma:
```
acumula(lista, f_acumulacion, valor_inicial)
```
La función de acumulación tendrá dos parámetros, a saber:
```
f_acumulacion(acumulado, siguiente)
```
donde 
- acumulado: es el valor que estamos acumulando
- siguiente: el el valor del item de la lista que estamos procesando

Además
- valor_inicial: será el valor de acumulación del que partimos. En el caso de que la función sea el producto, el valor inicial será 1, en el caso de la suma será cero, si queremos procesar una cadena el valor inicial sería "".

Mira la ayuda a la implementación para verlo con un ejemplo.

## Ayuda a la implementación

Si buscamos el producto de la lista `[2, 4, 6, 8, 10]`

1. La llamada a la función será `acumula([2, 4, 6, 8, 10], producto, 1)
2. La función producto tendrá esta firma 
```
producto(acum, siguiente)
```
3. La primera llamada será así y devolverá
```
producto(valor_inicial, lista[0])
producto(1, 2) -> 2
```
4. En el siguiente paso
```
producto(acum, lista[1])
producto(2, 4) -> 8
```
5. En el siguiente paso
```
producto(acum, lista[2])
producto(8, 6) -> 48
```

Y así hasta el final de modo que acum tendrá el valor 3840
