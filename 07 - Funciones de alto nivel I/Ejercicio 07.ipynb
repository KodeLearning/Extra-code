# Ejercicio 7. Cifrado y descifrado ad hoc

## Enunciado

Con lo aprendido ahora y utilizando lambdas debes crear una función `cifra(mensaje)` que aplique al mensaje las siguientes transformaciones.
1. Pasar el mensaje a MAYUSCULAS 
2. Cifrar el mensaje con Cesar + 4
3. Sustituir las letras A por un *

Y crea otra función llamada `descifra(mensaje)` que haga los pasos contrarios en el orden adecuado (excepto el de mayusculas, con el que invariablemente hemos perdido información)

1. Sustituir * por A
2. Cifrar el mensaje con Cesar - 4

Y comprueba que funciona

## Ayuda a la implementación

Recuerda divide y vencerás.

1. Crea una funcion `to_mays` que sea un alias de modifica a mayusculas
2. Crea una funcion `cesar4` que sea un alias de cesar con la clave 4
3. Crea una funcion `subsA` que sea un alias de modifica sustituyendo las Aes a asteriscos
4. Crea la función buscada por composición de funciones

Identifica tú los pasos del descifrado y sigue por ahí
