# Ejercicio 10. Normalizar una cadena

## Enunciado

Se trata de dejar las cadenas de entrada a nuestro cifrador de forma que solo incluyan letras del alfabeto (ABCDEFGHIJKLMNÑOPQRSTUVWXYZ) espacios y signos de puntuación.

Se evitarán las letras con tildes ` ´ o diéresis ¨.

Además se pretende que siempre lleguen al cifrador mayúsculas.

Construye un normalizador que sea la composición de las funciones necesarias que primero cambien las letras raras a su forma normal y después lo pase a mayúsculas. Ni espacios, ni signos de puntuación deben perderse
