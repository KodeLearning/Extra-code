# Ejercicio 6. Acumulado de listas con lambdas

## Enunciado

Partiendo de la función de acumulado del ejercicio 3 de este bloque, debes hacer:

1. Partiendo de la lista de la compra del ejercicio anterior. Calcular el total de la compra con la función acumulado.
```
productos = [("Camiseta", 15.99, 2),
             ("Pantalón", 25.50, 1),
             ("Zapatos", 50.00, 1),
             ("Reloj", 70.25, 3),
             ("Gorra", 10.00, 2),
             ("Bolso", 35.99, 1),
             ("Bufanda", 12.50, 4),
             ("Chaqueta", 89.99, 1)]
```

Recuerda que el precio de cada item es el precio unitario, elemento 1 de la tupla, multiplicado por el número de unidades, elemento 2 de la tupla.

2. Partiendo de una lista de cadenas concatenarlas todas separadas por el caracter '-'

3. Encontrar la palabra más larga de esa misma lista.

4. Encontrar la diferencia entre el número más grande de una lista y el más pequeño



## Ayuda a la implementación

Para evitar el guión inicial del ejemplo 2. Puedes hacer que en la frase acumula el tercer parámetro `valor_inicial` sea un valor por defecto. Si se informa se procesa la lista desde el inicio al final, si no se informa, el valor inicial es el primer item (indice 0) de la lista y se procesa el resto de la lista.

Para el ejemplo 4 lo mejor es encontrar el más grande, luego el más pequeño y obtener la resta
