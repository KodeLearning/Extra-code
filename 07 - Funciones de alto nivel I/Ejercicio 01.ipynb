# Ejercicio 1. Transformando una cadena

## Enunciado

Debes construir una función que permita procesar todos los caracteres de una cadena de forma que sean sustituidos por otro.

La sustituciones que buscamos son:
  1. Convertir cualquier consonante en su mayúscula
  2. Eliminar todas las ocurrencias de vocales (minúsculas o mayúsculas)
  3. Aplicar cifrado cesar clásico, es decir, con clave 1 (Ejercicio 2 de bucles infinitos)
  4. Convertir en jerigonza (Ejercicio 6 de secuencias)


## Ayuda a la implementación

Básicamente debes construir una función que pida una cadena y su función de modificación. La función de modificación es una función sin parámetros. Es decir, debe invocarse así:
```
modifica(cadena, funcion_modificadora)
```
