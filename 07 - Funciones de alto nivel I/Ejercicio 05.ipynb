# Ejercicio 5. Filtrado de listas con lambdas

## Enunciado

Partiendo de la función de filtrado del ejercicio 2 de este bloque, utiliza la función filtrar para:

1. Extraer todas las palabras que empiezan por vocal de la siguiente lista `["Destino", "Inmortal", "", "Camino", "Aventura", "Exótico", "Atemporal", "deslumbrante", "apasionado", "Fascinante", "Misterioso"]`
2. Todos los números que al ser divididos por 6 dan de resto 2 de una lista de numeros cualquiera
3. Todos los aprobados (nota mayor o igual a 5) de la siguiente lista de notas de programacion 101
```
[('Juan Perez', 9), ('Maria Rodriguez', 5), ('Luis Martinez', 1), ('Ana Garcia', 5), ('Pedro Sanchez', 8), ('Laura Gomez', 10), ('Carlos Hernandez', 5), ('Lucia Fernandez', 3), ('Miguel Diaz', 9), ('Adriana Torres', 5), ('Diego Vazquez', 2)]
```
4. Partiendo de la siguiente lista de productos de una compra
```
productos = [("Camiseta", 15.99, 2),
             ("Pantalón", 25.50, 1),
             ("Zapatos", 50.00, 1),
             ("Reloj", 70.25, 3),
             ("Gorra", 10.00, 2),
             ("Bolso", 35.99, 1),
             ("Bufanda", 12.50, 4),
             ("Chaqueta", 89.99, 1)]
```

donde el primer elemento de cada tupla es el nombre del producto, el segundo el precio unitario y el tercero la cantidad comprada. Partiendo de dicha lista filtrar los productos cuyo precio total es menor a 50. No requiere pin de la tarjeta de crédito.
