# Ejercicio 4. Personajes de rol

## Enunciado

Estás creando un programa que te ayude a ser el DM de una partida de rol.

1. Define una **clase** `Personaje` que permita crear personajes con sus características principales. La clase debe tener los siguientes atributos:

- **nombre**: Nombre del personaje
- **fuerza**: Número entre 0 y 20
- **destreza**: Número entre 0 y 20
- **raza**: una cadena con uno de los valores "elfo", "enano", "gnomo" o "humano"
- **jugador**: Nombre del jugador al que pertenece el personaje

  Puedes crear un personaje solamente sabiendo el nombre, la raza y el jugador. Por defecto, la fuerza y la destreza serán cero.

2. Crea un _método_ `generar` que permita establecer un valor para la fuerza y la destreza de forma aleatoria entre 3 y 17.

3. Crea un _método_ que permita realizar un combate entre dos personajes. El combate se funciona de la siguiente manera:

  1. El personajeA _ataca_ al personajeB, se invoca el método `ataca` del personajeA.
  2. Se lanza un dado de 20 (es decir, obtienes un número aleatorio entre 1 y 20).
    1. Si sale 20 el personajeA gana automáticamente. 
    2. Si sale 1 el personajeB gana automáticamente.
    3. En otro caso, se suma el valor de la fuerza y la destreza al valor obtenido. Esta será la puntuación del personajeA
    4. Se lanza de nuevo un dado de 20 y el valor se suma con la fuerza y la destreza del personajeB.
  3. Gana el personaje con mayor puntuación. Si ambas puntuaciones son iguales, pierde el atacante.
  4. El método devuelve `True` si gana el personajeA y `False` si gana el personajeB.

4. Crea un personaje Elfo para ti de nombre _Angrod_ y establece la fuerza y la destreza de forma aleatoria
5. Crea un personaje Enano para tu mejor amigo llamado _Brokk_ y establece su fuerza en 15 y la destreza en 12
6. Haz que el elfo ataque al enano ¿quién gana?



## Ayuda a la implementación

Números aleatorios: utiliza la función `random.randint(a,b)` para generar un números enteros aleatorios entre dos valores dados. Ref. https://docs.python.org/3/library/random.html?highlight=randint#random.randint
