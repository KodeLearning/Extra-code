# Ejercicio 2. Triangulo

## Enunciado

Crea una clase `Triangulo` que:

1. Se construya a partir del valor de sus tres ángulos
2. Los ángulos sean accesibles como atributos
3. Tenga un método `comprobar_angulos` que devuelva `True` si los ángulos son correctos (suman 180 grados) y `False` en caso contrario

## Ayuda a la implementación

Se puede crear un objeto triángulo pasando al constructor tres ángulos cualesquiera. Luego se comprobará si el triángulo es válido con un método.

Por ejemplo, el triángulo `t = Triangulo(20,20,20)` no es un triángulo válido pero el triángulo `r = Triangulo(60, 60, 60)` sí lo es.
