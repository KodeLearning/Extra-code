# Ejercicio 8. Películas. Versión mejorada

## Enunciado

Vamos a utilizar la clase `Show` del ejercicio 3 para mejorarla y permitirnos mantener los datos de dónde puede verse y de las cartelera de algunos cines. Para ello debes implementar las siguientes tareas:

1. Crea la **clase** `Cine` para guardar la información de un cine que consiste en su nombre y su dirección. Además, debe tener un **atributo de instancia** para guardar los pases disponibles en una lista que, inicialmente estará vacía.

2. Agrega a la clase `Cine` una implementación del método mágico `__str__` para que devuelva el nombre del cine y facilite la impresión de los datos.

3. Crea la **clase** `Pelicula` que herede de la clase `Show`. El tipo siempre será `'film'`, por lo que no será un parámetro de entrada en el constructor. Además, debe tener un **atributo de instancia** para guardar los pases disponibles en una lista que, inicialmente estará vacía.

4. Crea la **clase** `Pase` que guarde los datos de cine, película y hora. La hora será simplemente una cadena como `'17:45'`. Cine y película deben ser una instancia de la clase `Cine` y una instancia de la clase `Pelicula` respectivamente.

5. Incluye una validación en el constructor de los tipos de cine y película.

6. Al crear un pase se debe actualizar la lista de los pases del cine y de la película correspondientes con el **objeto** (pase) recién creado.

7. Agrega el método `ver_cartelera` en la clase `Cine` que muestre las películas que pueden verse y a qué horas (información de los pases)

8. Agrega el método `donde_la_veo` en la clase `Pelicula` que muestre los cines en los que se puede ver la película y a qué horas (información de los pases)

## Ayuda a la implementación

Recuerda que `self` hace referencia al **objeto** instanciado dentro del código de la **clase**.

Para marcar los errores utiliza la excepción `ValueError`
