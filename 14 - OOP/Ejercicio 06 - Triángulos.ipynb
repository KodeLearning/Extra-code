# Ejercicio 6. Triángulos

## Enunciado

Crea una clase `Triangulo` que:

1. Se construya a partir del valor de la longitud sus tres lados en centímetros.
2. Tenga un método `validar` que compruebe que esos tres valores forman un triángulo.
3. Tenga un método `get_tipo` que devuelva el tipo de triángulo según sus lados. Inicialmente el resultado será `'indefinido'`.
4. Crea una clase para cada tipo de triángulo según sus lados que herede de la clase `Triangulo` de forma que:
  1. La clase `Equilatero` solamente aceptará un lado para crearlo y el método `get_tipo` devolverá como resultado el valor `'equilátero'`.
  2. La clase `Isosceles` solamente aceptará dos valores en el constructor. El primero de ellos corresponderá a la longitud de los dos lados iguales en centímetros y el segundo al tercer lado. El método `get_tipo` devolverá la cadena `'isósceles'`.
  3. La clase `Escaleno` se construye con el valor de los tres lados y el método `get_tipo` devolverá el valor `'escaleno'`.
5. Instanciar tres triángulos, uno de cada tipo y, para cada uno de ellos, utilizar el método `validar` para comprobar que los valores utilizados permiten la creación de un triángulo válido.
6. Para cada uno de los triángulos anteriores, comprobar el resultado del método `get_tipo`.
7. Utilizar el método `isinstance` para comprobar que los tres triángulos anteriores son de tipo `Triangulo` además del tipo específico correspondiente (`Equilatero`, `Isosceles` o `Escaleno`).
8. Por último, comprobar que la creación de un triángulo con los valores 1, 2, 4 no genera un triángulo válido.


## Ayuda a la implementación

Para validar si un triángulo es válido a partir de la longitud de sus lados, lo único que debes hacer es usar el **Teorema de la desigualdad del triángulo**, el cual establece que la suma de dos lados del triángulo siempre es mayor que la medida del tercer lado. Si esto resulta ser verdad para las tres combinaciones posibles de las sumas, entonces tienes un triángulo. Es decir, si los lados son _a_, _b_ y _c_ se tiene que cumplir que:

- a + b > c
- a + c > b
- b + c > a

Revisa la documentación de la función `isinstance` en la documentación oficial de Python: https://docs.python.org/es/3/library/functions.html#isinstance
