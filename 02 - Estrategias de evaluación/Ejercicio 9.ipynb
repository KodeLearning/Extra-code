# Ejercicio 9. Fibonacci y razón aurea, sin iteraciones
## Enunciado

Un ejercicio clásico de iteraciones es crear la sucesión de fibonacci. Pero hay una manera de calcular cualquier término de la sucesión con una fórmula

$x_{n} = \frac{\varphi ^{n} - (1 - \varphi )^{n}}{\sqrt{5}}$

Donde $\varphi$ es el número aúreo cuyo valor es 1.618\033988749895

Dado que al ser irracional perdemos precisión el cálculo debe ser redondeado al entero más próximo.

Crea una función `fibo(n)` que devuelva el término de fibonacci en la posición n.

## Ayuda a la implementación

Utiliza el valor de $\varphi$ como constante.

Para redondear al entero más próximo, un buen reto es hacer tú la función de redondeo
