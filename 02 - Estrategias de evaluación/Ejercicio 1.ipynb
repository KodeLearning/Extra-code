# Ejercicio 1. Sacar dinero de un cajero automático

## Enunciado

Debes crear una función que simule sacar dinero de un cajero automático. Tendrás una variable llamada `saldo_cuenta` que contendrá el saldo de la cuenta del usuario. Esta variable estará inicializada con el valor que tú quieras y se definirá fuera de la función.

Debes crear una función llamada `sacar_dinero` a la que informarás la cantidad que el usuario quiere sacar.

Si la cantidad es mayor que el saldo, la función debe devolver 0. Si la cantidad es menor que el saldo, la función devolverá el importe extraido (simula el cajero) y la variable `saldo_cuenta` debe actualizarse en consecuencia.



## Ayuda a la implementación

Recuerda que los parámetros numéricos de la función (puedes poner tantos como necesites aunque uno, necesariamente, sea la cantidad a sacar) se pasan por valor.
