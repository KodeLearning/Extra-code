# Ejercicio 7. Animales

## Enunciado

Crea una estructura de datos usando listas para almacenar en una sola variable esta clasificación de especies animales:

Clasificación animal

Agrega una rama para las aves, rapaces, que incluya al águila y al azor.

Crea una función `clasificar` que acepte como parámetro el nombre de un animal y muestre su clasificación. Por ejemplo, "Cobra" mostrará la clasificación "Reptil Serpiente". 

## Ayuda a la implementación

Para cada nodo puedes usar una lista que tenga el nombre guardado como primer elemento y como segundo elemento una lista con los _hijos_.

Recuerda que puedes usar el operador `in` para comprobar si un elemento pertenece a una lista, pero ten en cuenta que en Python "Cobra" no es lo mismo que "cobra".
