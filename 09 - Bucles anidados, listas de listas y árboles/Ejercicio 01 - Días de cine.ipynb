# Ejercicio 1. Días de cine

## Enunciado

Imagina que quieres llevar la cuenta de los días que vas al cine cada mes. Para cada mes, si vas al cine un día, anotas el día correspondiente.

Crea una estructura de datos llamada `dias_de_cine` que te permita anotar los días que vas al cine utilizando listas anidadas. Cada elemento de la lista será a su vez una lista que tendrá por elementos los días que has ido al cine ese mes.

Así, si el mes de enero, has ido al cine los días 1 y 6, el primer elemento de la lista `dias_de_cine` será una lista con los números `[1, 6]`.

Si un mes no has ido al cine, la lista de días correspondiente a ese mes será una lista vacía.

Crea la lista para almacenar los días de enero a mayo:

- Enero: 1, 6
- Febrero: 15
- Marzo: 7, 17, 23, 29
- Abril: no has ido al cine
- Mayo: 1, 10, 20

¿Cómo calcularías el número máximo de días que has ido al cine un mes? Según los datos, el resultado es 4. Imprime el resultado utilizando la función `print`.

## Ayuda a la implementación

Recuerda que la longitud de una lista se puede obtener con la función `len`.
