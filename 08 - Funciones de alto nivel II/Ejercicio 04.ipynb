# Ejercicio 4. Devolviendo datos con una clousure

## Enunciado

Con una función que devuelve funciones debes crear un paginador de datos.

El creador da paginadores recibirá como parámetros la longitud de la página n (es decir, cuantos items ha de devolver en cada llamada al paginador) y la lista de datos a paginar.

Deberá devolver una función que cada vez que sea llamada retornará los n siguientes elementos que no hayan sido mostrados. Si llega al final de la lista devolverá n o menos y después simplemente una lista vacía.

Sirva como ejemplo de la funcionalidad que buscamos:
```
>>> paginador = crea_paginador(2, [2, 3, 4, 5, 7, 9, 11])
>>> paginador()
[2, 3]
>>> paginador()
[4, 5]
>>> paginador()
[7, 9]
>>> paginador()
[11]
>>> paginador()
[]
>>> paginador()
[]
```
