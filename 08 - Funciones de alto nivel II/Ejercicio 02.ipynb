# Ejercicio 2. Sumadores de números y de letras

## Enunciado

Usando funciones que devuelven funciones crea

1. Una función que cree sumadores tal que:
```
>>> suma2 = crea_sumador(2)
>>> suma2(3)
5
```

Y que el programador pueda crear cualquier sumador con cualquier valor

2. una función que cree sufijadores tal que:

```
>>> suma_ote = crea_sufijador("ote")
>>> suma_ote("taza")
tazote
>>> suma_ote("camión")
camionote
```

Es decir que si se aplica la función a una cadena acabada en vocal, le quite la última vocal y le añada el sufijo indicado al crearlo

Y si se aplica la función a una cadena acabada en consonante, simplemente le añada al final el sufijo
