# Ejercicio 1. Constructor de jerigonzas

## Enunciado

Ya sabes construir una función que transforme una frase en jerigonza. 

Debes construir ahora una función a la que le informes como parámetros de entrada la sílaba que quieres poner, no necesariamente debe ser `pi`. 

Es decir, debes construir una función tal que  al hacer 

```
`>>> jerigonza_pe = crea_jerigonza('pe')
```

luego puedas usar jerigonza_pe de este modo

```
>>> jerigonza_pe("Hola")
'Hopelape'
```

## Ayuda a la implementación

Debes utilizar una función de nivel superior que reciba la sílaba y que devuelva la función jerigonza que ya usaste con dicha sílaba en lugar de la origina `pi`
