# Ejercicio 1. Acceso por pin

## Enunciado

Debes crear un programa que pida un pin de acceso de 4 cifras y solo permita el acceso cuando se adivine el pin.




## Ayuda a la implementación

- Recuerda que input siempre devuelve strings aunque se hayan escrito números.


```python
from typing import Any

def check_pin(pin: Any) -> bool:
  real_pin = 1234
  has_access = False

  if pin == real_pin:
    has_access = True

  return has_access
```

```python
def get_access():
  fail = 0

  while True:
    pin = input('Introduce tu pin para continuar:')

    if check_pin(int(pin)):
      print('Acceso concedido.')
      break
    else:
      if fail > 2:
        print('Demasiados intentos, ¡vuelve otro día!')
        break
      else:
        print('No has conseguido acceso.')
        fail += 1
```

```python
get_access()
```
