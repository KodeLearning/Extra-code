# Ejercicio 1. Conversor de temperaturas.

## Enunciado

**Parte 1**. Crea una función que convierta grados centígrados en grados Fahrenheit.

La fórmula de conversión es la siguiente:

$$
ºF =  \frac{9}{5} · ºC + 32
$$

**Parte 2**. Crea la función inversa, convertir grados Fahrenheit a Celsius

En este caso la fórmula es:

$$
ºC = \frac{ºF - 32}{1.8}
$$

```python
def celsius_to_fahrenheit(grades):
  return 9 / 5 * grades + 32
```

```python
def fahrenheit_to_celsius(grades):
  return (grades - 32) / 1.8
```
