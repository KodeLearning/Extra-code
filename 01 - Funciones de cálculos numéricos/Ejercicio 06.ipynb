## Ayuda a la implementación

Recuerda que la composición de funciones es que la a la salida de una función se le aplica otra. Algo así:

```python
def f(x):
  return x * 2

def g(x):
  return x - 1
```

Si f está compuesta con g significa que calculamos g(f(x)) cuyo resultado es 2x - 1

Si g está compuesta con f significa que calculamos f(g(x)) cuyo resultado es 2(x-1)

Compruébalo

```python
g(f(3))
```

```python
f(g(3))
```
