# Ejercicio 2. Conversor de segundos.

## Enunciado

**Parte 1.** Crea una función que devuelva las horas, minutos y segundos de un tiempo dado en segundos.

**Parte 2.** Crea la función contraria, indicándole el número de horas, minutos y segundos que te de el total de segundos



## Ayuda a la implementación

Recuerda que un minuto tiene 60 segundos, una hora tiene 60 minutos o lo que es lo mismo 3600 segundos.

El resultado puedes devolverlo así.

```
return horas, minutos, segundos
```

````python
def seconds_to_time(time_in_seconds: int) -> str:
    hours = time_in_seconds // 3600
    minutes = (time_in_seconds % 3600) // 60
    seconds = (time_in_seconds % 3600) % 60
    print (f'{hours} horas, {minutes} minutos y {seconds} segundos')
    return str(hours) + ', ' + str(minutes) + ', ' + str(seconds)
````

````python
seconds_to_time(723478238)
````
